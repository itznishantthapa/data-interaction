```python
ds=pd.read_csv('../socialNetworkAds.csv')
ds.shape
```

```python
ds.drop(columns=['User ID','Gender'],inplace=True)
ds.head(3)
```

```python
ds.columns
```

```python
sns.scatterplot(x='Age',y='EstimatedSalary',data=ds,hue='Purchased')
plt.show()
```

```python
x=ds.iloc[:,:-1]
y=ds['Purchased']
```

```python
from numpy.random import rand
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr2=LogisticRegression()
lr2.fit(x_train,y_train)
```

```python
lr2.score(x_test,y_test)*100
```

```python
lr2.predict([[32,150000]])
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr2)
plt.show()
```
